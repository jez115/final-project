# Playground Dataset

We are measuring the number of playgrounds in each neighborhood and consider neighborhoods with **more** playgrounds to be "**better**." The data was last updated in 2021 because of a broken data feed.

Here is the link to the dataset: https://data.wprdc.org/dataset/playgrounds/resource/47350364-44a8-4d15-b6e0-5f79ddff9367

Below is a preview of the original data:

```python
import pandas as pd
import numpy as np
%matplotlib inline
playground = pd.read_csv('playground.csv')
playground.head(10)
```

By organizing the neighborhoods in alphabetical order, we can see how many playgrounds correspond to the same neighborhood:

```python
p2 = playground.sort_values(by = "neighborhood")
p2.head(10)
```

Below is condensed dataset of the neighborhoods and their playground count:

```python
p2 = playground['neighborhood'].value_counts().sort_values(ascending = False)
p2.head(10)
```

Finally, we can summarize the data visually with a bar chart.

```python
p2.plot(kind = "bar", width = .4, figsize=(25,10), ylim=(0,8))
plt.ylabel("number of playgrounds")
plt.show()
```
