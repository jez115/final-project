The metric used is fire incidents reported per neighborhood. The neighborhood with the least number of fires is the best. The dataset that provided this information was Fire Incidents in City of Pittsburgh (https://data.wprdc.org/datastore/dump/8d76ac6b-5ae8-4428-82a4-043130d17b02). 

```python
import pandas as pd
import matplotlib.pyplot as plt

fire = pd.read_csv('fire incidents.csv')
fire = fire.dropna()
print("the following is a sample of what each row looks like in the dataset\n")
fire.head(10)
```

```python
fire["neighborhood"].value_counts().sort_values(ascending= True).plot.bar(title = "Number of Fire Incidents in Pittsburgh by Neighborhood", 
                                                              xlabel="Neighborhood Names", ylabel="Number of fire incidents", 
                                                              figsize=(18, 8),fontsize = 8, rot=990)
```

According to the bar graph, the ranking for best neighborhoods goes as following: 
1. Regent Square
2. Ridgemont
3. East Carnegie
