```python
num_unique_items = trash['neighborhood'].nunique()
unique_items = trash['neighborhood'].unique()

print(f"There are {num_unique_items} unique items.")
print(unique_items)
```

```python
my_dict = dict()
for index, row in trash.iterrows():
    for column in trash:
        if (column == 'neighborhood'):
            if (row['neighborhood'] in my_dict):
                my_dict[row['neighborhood']] += 1
            else:
                my_dict[row['neighborhood']] = 1
print(my_dict)
```
