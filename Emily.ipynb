The metric is criminal activity

dataset counts crime incidents between 2024-01-01 and 2025-03-31


```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
crimes=pd.read_csv("criminal_activity.csv", dtype={"Tract":str})
crimes.head()
```

```python
crimes.groupby("Neighborhood").count()
```

```python

plt.figure(figsize=(18, 6)) 
crime_counts=crimes["Neighborhood"].value_counts().sort_values()
plt.title("Crime in Pittsburgh Neighborhoods")
plt.ylabel("Number of Crimes Reported")
crime_counts.plot(kind='bar')


plt.show()
```

**best neighborhoods:**
1. Saint Clair
2. Ridgemont
