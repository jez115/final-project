The following charts summarize this:

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

crimes=pd.read_csv("criminal_activity.csv", dtype={"Tract":str})

plt.figure(figsize=(18, 6)) 
crime_counts=crimes["Neighborhood"].value_counts().sort_values()
plt.title("Crime in Pittsburgh Neighborhoods")
plt.ylabel("Number of Crimes Reported")
crime_counts.plot(kind='bar')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

fire = pd.read_csv('fire incidents.csv')
fire = fire.dropna()

fire["neighborhood"].value_counts().sort_values(ascending= True).plot.bar(title = "Number of Fire Incidents in Pittsburgh by Neighborhood", 
                                                              xlabel="Neighborhood Names", ylabel="Number of fire incidents", 
                                                              figsize=(18, 8),fontsize = 8, rot=990)
plt.show()
```
